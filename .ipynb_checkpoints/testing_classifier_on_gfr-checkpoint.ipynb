```python
torch.cuda.empty_cache()
```

```python
def collate_fn(batch):
    batch = list(filter(lambda x: x is not None, batch))
    return torch.utils.data.dataloader.default_collate(batch)
```

```python
prefix='test'
tests = ['test.csv',
         prefix+'_over_60.csv',
         prefix+'_30_59.csv',
         prefix+'_15_29.csv',
         prefix+'_under_15.csv']
```

```python
test_csv = '/workspace/data/drives/Local_SSD/sdc/kidney_disease/ICD_ECG_GFR/all_gfr_with_binaries_cleaned.csv'
root = '/workspace/data/drives/Local_SSD/sdd/data/Remade with New Coefficents'
model_origin = 'RCRI_Net'
target='over_60'
additional_inputs = []
```

```python
test_ds = EchoECG(root=root,
                  csv=test_csv,
                  model=model_origin, 
                  rolling=0, 
                  downsample=1,
                  target=target, 
                  one_lead=False,)
                  #,additional_inputs=additional_inputs)
print(len(test_ds)) 
bs = 2000
test_dataloader=torch.utils.data.DataLoader(test_ds,
                                            batch_size=bs, 
                                            num_workers=40, 
                                            drop_last=False, 
                                            collate_fn=collate_fn)
```

```python
model = models.EffNet(channels = [32,16,24,40,80,112,192,320,1280,1],dilation = 2,
                                  stride = 8,
                                  reg = False, 
                                  start_channels=12)
                                 #,num_additional_features=2)
```

```python
model.load_state_dict(torch.load('/workspace/kai/phecode/Training/Standard EKG to all stages CKD/12LNA_all_stages/best_auc_2.pt'))
```

```python
all_labels = []
all_preds = []
with torch.no_grad():
    for ecg, labels in tqdm(test_dataloader):
        
        all_preds += list(model(ecg))
        all_labels += list(labels)
        
```

```python
len(all_preds)
```

tensor_preds = torch.tensor(all_preds)
tensor_labels = torch.tensor(all_labels)

mae = sklearn.metrics.mean_absolute_error(all_labels, all_preds)
r2 = sklearn.metrics.r2_score(all_labels, all_preds)
print(mae*250, r2)

```python
cm=sklearn.metrics.confusion_matrix(all_labels, [x>0.5 for x in all_preds])
sklearn.metrics.ConfusionMatrixDisplay(cm).plot()

fpr, tpr, thresholds = sklearn.metrics.roc_curve(all_labels, all_preds)
sklearn.metrics.RocCurveDisplay(fpr=fpr, tpr=tpr).plot()
print(sklearn.metrics.auc(fpr, tpr))
```
