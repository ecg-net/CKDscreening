```python
import numpy as np

from paths import FilePaths
from tuningfunctions import get_data_loaders, get_criterion, Objective, run_trials
import models
import torch
import pandas as pd
import matplotlib.pyplot as plt
from ECG import EchoECG
from constants import MODEL_NAME, OPTIMIZERS, MAX_EPOCHS, METRIC, MIN_LR, MAX_LR, PATIENCE, SCHEDULER, STEP, GAMMA, PRUNER, NUM_TRIALS, DIRECTION
import wandb
from sklearn import metrics
```

```python
Train_csv = '/workspace/data/drives/Local_SSD/sdc/kidney_disease/stage_6_to_3_ESRD_ICD_9_10_to_EKG_train.csv'
Val_csv = '/workspace/data/drives/Local_SSD/sdc/kidney_disease/stage_6_to_3_ESRD_ICD_9_10_to_EKG_val.csv'
root = '/workspace/data/drives/Local_SSD/sdd/data/Remade with New Coefficents'
model_origin = 'RCRI_Net'
Project_name = 'PreOpNet ECG to CKD ICD9'
target='label'

train_ds = EchoECG(root=root, csv=Train_csv, model=model_origin ,rolling=0, downsample=1, target=target)
test_ds = EchoECG(root=root, csv=Val_csv, model=model_origin, rolling=0, downsample=1, target=target)
print(len(train_ds),len(test_ds)) 
bs = 2000
train_dataloader = torch.utils.data.DataLoader(train_ds, batch_size=bs, shuffle=True, drop_last=False)
val_dataloader=torch.utils.data.DataLoader(test_ds, batch_size=bs, drop_last=False)
```

```python
inputs, labels = next(iter(val_dataloader))
```

```python
model = models.EffNet(depth = [1,2,2,3,3,3,3],
                      channels = [32,16,24,40,80,112,192,320,1280,1],
                      dilation = 2,
                      stride = 8,
                      reg=False)
model.load_state_dict(torch.load('/workspace/kai/phecode/Training/PreOpNet ECG to CKD ICD9/Binary CKD all stages/best_auc.pt'))
model.eval()
```

```python
with torch.no_grad():
    outputs = model(inputs)
```

```python
true_labels = labels[labels==1]
true_outputs = outputs[labels==1]

print(torch.sum(true_labels), torch.sum(true_outputs))
```

```python
fpr, tpr, thresholds = metrics.roc_curve(labels, outputs)
roc_auc = metrics.auc(fpr, tpr)
display = metrics.RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc)
display.plot()
```
