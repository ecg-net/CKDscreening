```python
torch.cuda.empty_cache()
```

```python
def collate_fn(batch):
    batch = list(filter(lambda x: x is not None, batch))
    return torch.utils.data.dataloader.default_collate(batch)
```

```python
test_csv = '/workspace/data/drives/Local_SSD/sdc/kidney_disease/DefinitiveAllStagesData/all_stages_non_train.csv'
root = '/workspace/data/drives/Local_SSD/sdd/data/Remade with New Coefficents'
model_origin = 'RCRI_Net'
target='label'
additional_inputs = []
```

```python
model = models.EffNet(channels = [32,16,24,40,80,112,192,320,1280,1],dilation = 2,
                                  stride = 8,
                                  reg = True, 
                                  start_channels=12)
                                 #,num_additional_features=2)
```

```python
#NOTE: 12LNA_normalized_gfr contains a model that was trained on *all* available eGFR data
```

```python
model.load_state_dict(torch.load('/workspace/kai/phecode/Training/standard_ecg_gfr/12LNA_normalized_gfr/best_loss_model_192_val_loss=-0.0134.pt'))
```

```python
test_ds = EchoECG(root=root,
                  csv=test_csv,
                  model=model_origin, 
                  rolling=0, 
                  downsample=1,
                  target=target, 
                  one_lead=False,
                  return_filename=True)
                  #,additional_inputs=additional_inputs)
bs = 2000
test_dataloader=torch.utils.data.DataLoader(test_ds,
                                            batch_size=bs, 
                                            num_workers=40, 
                                            drop_last=False, 
                                            collate_fn=collate_fn)


all_labels = []
all_preds = []
all_filenames=[]
with torch.no_grad():
    for ecg, labels, fnames in tqdm(test_dataloader):
        
        all_filenames = [*all_filenames, *fnames]
        all_preds += list(model(ecg))
        all_labels += list(labels)
        
```
