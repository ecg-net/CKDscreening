```python
import numpy as np

from paths import FilePaths
from tuningfunctions import get_data_loaders, get_criterion, Objective, run_trials
import models
import torch
import pandas as pd
import matplotlib.pyplot as plt
from ECG import EchoECG
from constants import MODEL_NAME, OPTIMIZERS, MAX_EPOCHS, METRIC, MIN_LR, MAX_LR, PATIENCE, SCHEDULER, STEP, GAMMA, PRUNER, NUM_TRIALS, DIRECTION
import wandb
from tqdm import tqdm
import sklearn
```

```python
torch.cuda.empty_cache()
```

```python
Train_csv = '/workspace/data/drives/Local_SSD/sdc/kidney_disease/DefinitiveAllStagesData/all_stages_train.csv'
Val_csv = '/workspace/data/drives/Local_SSD/sdc/kidney_disease/DefinitiveAllStagesData/all_stages_val.csv'
root = '/workspace/data/drives/Local_SSD/sdd/data/Remade with New Coefficents'
model_origin = 'RCRI_Net'
Project_name = 'CLASSIFIER_DEFINITIVE_ONE_LEAD'
target='label'
```

```python
def collate_fn(batch):
    batch = list(filter(lambda x: x is not None, batch))
    return torch.utils.data.dataloader.default_collate(batch)
```

```python
target_list = ['label']
torch.cuda.empty_cache()
root_list = {
    #'1 heartbeat': '/workspace/data/drives/Internal_SSD/sdd/data/Limited Datasets/2008-2020 EKG QRS Numpy',
    #'P Include': '/workspace/data/drives/Internal_SSD/sdd/data/Limited Datasets/P_include',
    #'P Exclude':'/workspace/data/drives/Internal_SSD/sdd/data/Limited Datasets/P_exclude',
    #'QRS Include':'/workspace/data/drives/Internal_SSD/sdd/data/Limited Datasets/QRS_include',
    #'QRS Exclude':'/workspace/data/drives/Internal_SSD/sdd/data/Limited Datasets/QRS_exclude',
    #'T Include':'/workspace/data/drives/Internal_SSD/sdb/data/Limited Datasets/T_include',
    #'T Exclude':'/workspace/data/drives/Internal_SSD/sdb/data/Limited Datasets/T_exclude'
}

for target in target_list:
    y_train=pd.read_csv(Train_csv)[target]
    criterion=torch.nn.BCELoss()
    print(y_train[0:5])
    
    train_ds = EchoECG(root = root, csv = Train_csv,model = model_origin,rolling = 0,downsample=1,target=target, one_lead=True)
    test_ds = EchoECG(root = root, csv = Val_csv,model = model_origin,rolling = 0,downsample=1,target=target, one_lead=True)
    bs = 2000
    train_dataloader = torch.utils.data.DataLoader(train_ds, batch_size=bs,num_workers=40,shuffle = True,drop_last=False,collate_fn=collate_fn)
    val_dataloader=torch.utils.data.DataLoader(test_ds, batch_size=bs,num_workers=40,drop_last=False,collate_fn=collate_fn)
    # train_dataloader, val_dataloader=get_data_loaders(X_train, X_test, y_train, y_test)


    ## Edit the capitalized general parameter inputs in constants.py. 
    # Model specific parameters are inputted in models.py
    objective=Objective(MODEL_NAME, criterion, train_dataloader, val_dataloader, OPTIMIZERS, 
                        0.0001, 0.0001,'roc_auc', 250, 100, SCHEDULER, STEP, 
                        0.98, Project_name,model_origin,stride = 8,
                        dilation = 2,channels = [32,16,24,40,80,112,192,320,1280,1],
                        category='classification', start_channels=1)




    project_name = Project_name

    run_trials(objective, PRUNER, 1, DIRECTION, project_name)                
```
