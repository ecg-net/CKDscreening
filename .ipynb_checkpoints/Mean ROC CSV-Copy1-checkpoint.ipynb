```python
def bootstrap(x,y):
    y_total,yhat_total = x,y
    aucs = []
    for i in range(0,40):
        choices = np.random.choice(range(0,len(yhat_total)),int(len(yhat_total)/2))
        fpr,tpr, _ = roc_curve(y_total[choices],yhat_total[choices])
        aucs.append(auc(fpr,tpr))
    lower_point = round(np.percentile(aucs,2.5),2)
    higher_point = round(np.percentile(aucs,97.5),2)
    fpr,tpr, _ = roc_curve(y_total,yhat_total)
    mean = auc(fpr,tpr)
    return mean, lower_point, higher_point
def produce_csv(folder,data_set,output,csv):
    model = models.EffNet(stride = 8, dilation = 2,channels = [32,16,24,40,80,112,192,320,1280,1867])
    checkpoint = torch.load(folder)
    to_load = {'model': model}
    Checkpoint.load_objects(to_load=to_load, checkpoint=checkpoint)
    model.eval()
    root = data_set
    
    ds = EchoECG(root = root, csv = csv,model = 'RCRI_Net',rolling = 0,downsample=1)
    bs = 300
    test_dataloader = torch.utils.data.DataLoader(ds, batch_size=bs,num_workers=40,shuffle = False,drop_last=False)
    y_total = torch.Tensor()
    yhat_total = torch.Tensor()
    for x,y in tqdm(test_dataloader):
        yhat = model(x)
        y_total = torch.cat((y_total,y),0)
        yhat_total = torch.cat((yhat_total,yhat.detach().cpu()),0)
    phenocode = pd.read_csv('/workspace/data/NAS/Cedars EKG Analysis/phecode_icd9_rolled.csv')
    mean,low,high = [],[],[]
    phe = []
    for i in tqdm(range(0,yhat_total.shape[1]-1)):
        phe.append(phenocode.PheCode.unique()[i])
        a,b,c = bootstrap(y_total[:,i],yhat_total[:,i])
        mean.append(a)
        low.append(b)
        high.append(c)
    df = pd.DataFrame({"Phenocode":phe,"Mean":mean,"Low":low,"High":high})
    df['positives'] = torch.sum(y_total,0)
    info = pd.read_csv("Phenocode Info.csv")
    info = info.groupby("phecode").first()
    shortcut = phenocode.groupby("PheCode").first()
    desc = {}
    for i in shortcut.index:
        desc[i] = shortcut.Phenotype[i]
    group_labels = {}
    for i in info.index:
        group_labels[i] = info.group[i]
    def label(x):
        if x in group_labels.keys():
            return group_labels[x]
        else:
            return 'Extra'

    df['Group Labels'] = df.Phenocode.apply(label)
    def label(x):
        if x in desc.keys():
            return desc[x]
        else:
            return 'Missing'
    df['Phenotype'] = df.Phenocode.apply(label)
    df.to_csv(output,index=False)
```

```python
folder = 'PhenoCode_with_PreOpNet/checkpointoptimizer: AdamW, lr: 0.004, /best_loss_model_2_val_loss=-0.0269.pt'
root = '/workspace/data/drives/Local_SSD/sdb/UKBB No Processing'
output = "results UK_BB 1 year raw.csv"
produce_csv(folder,root,output,'/workspace/John/UK_BB_1_year_Phecode.csv')
```
