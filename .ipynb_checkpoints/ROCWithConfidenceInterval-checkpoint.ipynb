```python
from sklearn.metrics import roc_auc_score, roc_curve, confusion_matrix
import matplotlib.pyplot as plt
```

```python
def get_confidence_interval(x, percent):
    sorted_array = np.array(x)
    sorted_array.sort()
    return [np.percentile(sorted_array, (100-percent)/2), np.percentile(sorted_array, 100-((100-percent)/2))]
    

def confidence_interval_bootstrap(labels, predictions, iterations):
    #base auc calculation (what's the auc of the model on the subset?)
    base_auc = roc_auc_score(labels, predictions)
    
    #base metric calculation (on the whole subset)
    base_fpr, base_tpr, base_thresholds = roc_curve(labels, predictions)
    base_best_idx = np.argmax(base_tpr-base_fpr) 
    base_best_threshold = base_thresholds[base_best_idx]
    
    base_tn, base_fp, base_fn, base_tp = confusion_matrix(labels, predictions>base_best_threshold).ravel()
    
    base_sens = base_tp / (base_tp+base_fn)
    base_spec = base_tn / (base_tn+base_fp)
    base_ppv = base_tp / (base_tp+base_fp)
    base_npv = base_tn / (base_tn+base_fn)
    
    length = len(labels)
    AUCs = []
    sens = []
    specs = []
    PPVs = []
    NPVs = []
    threshes = []
    
    for i in range(iterations):
        indices = np.random.randint(0, length, length)
        sampled_labels = labels.iloc[indices].copy()
        sampled_predictions = predictions.iloc[indices].copy()
        AUCs.append(roc_auc_score(sampled_labels, sampled_predictions))
        
        fpr, tpr, thresholds = roc_curve(labels, predictions)
        best_idx = np.argmax(tpr-fpr) 
        best_threshold = thresholds[best_idx]
        
        tn, fp, fn, tp = confusion_matrix(labels, predictions>best_threshold).ravel()
        sensitivity = tp / (tp+fn)
        specificity = tn / (tn+fp)
        ppv = tp / (tp+fp)
        npv = tn / (tn+fn)
        
        sens.append(sensitivity)
        specs.append(specificity)
        PPVs.append(ppv)
        NPVs.append(npv)
        threshes.append(best_threshold)
    
    lower_auc, upper_auc = get_confidence_interval(AUCs, 95)
    lower_sens, upper_sens = get_confidence_interval(sens, 99)
    lower_spec, upper_spec = get_confidence_interval(specs, 99)
    lower_ppv, upper_ppv = get_confidence_interval(PPVs, 99)
    lower_npv, upper_npv = get_confidence_interval(NPVs, 99)
    lower_threshold, upper_threshold = get_confidence_interval(threshes, 99)
    
    
    print(f"AUC: {round(base_auc, 3)} ({round(lower_auc,3)}-{round(upper_auc,3)})")
    print(f"Best Threshold: {round(base_best_threshold, 3)} ({round(lower_threshold,3)}-{round(upper_threshold,3)})")
    print(f"Sensitivity: {round(base_sens, 3)} ({round(lower_sens,3)}-{round(upper_sens,3)})")
    print(f"Specificity: {round(base_spec, 3)} ({round(lower_spec,3)}-{round(upper_spec,3)})")
    print(f"PPV: {round(base_ppv, 3)} ({round(lower_ppv,3)}-{round(upper_ppv,3)})")
    print(f"NPV: {round(base_npv, 3)} ({round(lower_npv,3)}-{round(upper_npv,3)})")
    
    
        
```

```python
preds['under_sixty'] = ~preds.over_sixty
preds['all'] = True
```

```python
targets = [['all','has_ckd'], 
           ['all','has_mild_ckd'], 
           ['all','has_moderate_ckd'],
           ['all','has_esrd'], 
           ['has_diabetes', 'has_ckd'],
           ['has_hypertension','has_ckd'], 
           ['over_sixty', 'has_ckd'], 
           ['is_male', 'has_ckd'], 
           ['under_sixty', 'has_ckd'],
           ['under_sixty', 'has_mild_ckd'],
           ['under_sixty', 'has_moderate_ckd'],
           ['under_sixty', 'has_esrd']]

for model in ['twelve_lead_prediction','one_lead_prediction']:
    for column in targets:
        mask = preds[column[0]].copy()
        target = column[1]
        labels = preds[mask].copy()
        labels = labels[target].copy()
        predictions = preds[mask].copy()
        predictions = predictions[model].copy()
        
        print('\n')
        print('\n', model, column)
        confidence_interval_bootstrap(labels, predictions, 500)
    
```
