```python
import numpy as np
from sklearn.metrics import roc_curve, auc
from paths import FilePaths
from tuningfunctions import get_data_loaders, get_criterion, Objective, run_trials
import models,old_models
import torch.nn as nn
import torch
from ignite.handlers import ModelCheckpoint, global_step_from_engine, Checkpoint, DiskSaver
from importlib import reload

from tqdm import tqdm
import torch.nn.functional as F
import os
from matplotlib import pyplot as plt
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
from pthflops import count_ops

import pandas as pd
from ECG import EchoECG
from constants import MODEL_NAME, OPTIMIZERS, MAX_EPOCHS, METRIC, MIN_LR, MAX_LR, PATIENCE, SCHEDULER, STEP, GAMMA, PRUNER, NUM_TRIALS, DIRECTION
def get_checkpoint(x):
        high = 0
        loc = ''
        for i in os.listdir(x):
            for k in os.listdir(os.path.join(x,i)):
                if (k[-3:]=='.pt') and ('roc' in k):
                    m = int(k.split('.')[1])
                    print(m,high)
                    if m > high:
                        high = m
                        loc = os.path.join(x,i,k)
        return loc
def get_model(model_name, depth = [1,2,2,3,3,3,3],channels = [32,16,24,40,80,112,192,320,1280],dilation = 1,stride = 2,expansion = 6,additional_inputs=0):
    model = None
    if model_name == 'Columbia':
        model = getattr(models, model_name)(inital_kernel_num = 32,dropout = 0.6,conv1kernel1 = 7,conv1kernel2 = 3)
    elif model_name == 'Lancet':
        model_type='lancet'
        is_2d=False
        class_name = 'hemoglobin'
        lr = 1e-3
        model_name = 'lancet'
        binary=False
        batch_norm=True
        drop_prob=0.
        conv_width=3
        train_delta=0
        first_layer_out_channels=None
        first_layer_kernel_size=None
        normalize_x="lead"
        binary_cutoff = None
        normalize_y=False
        from ecgnet import model
        reload(model)
        model = model.build_model(model_type, 1, 12 if is_2d else 12, is_2d=is_2d, binary=binary, batch_norm=batch_norm, default_conv_kernel=conv_width, drop_prob=drop_prob,
                                 binary_cutoff=binary_cutoff, first_layer_out_channels=first_layer_out_channels, first_layer_kernel_size=first_layer_kernel_size)
    elif model_name == 'mayo':
        model = models.Mayo_Net()
    elif model_name == 'mayo  mortality':
        model = models.Mayo_Net_mortality()
    elif model_name == 'efficency':
        model = models.EffNet(depth = depth,channels = channels,dilation = dilation,stride = stride,expansion = expansion,num_additional_features=additional_inputs)
    if model_name == 'Old Columbia':
        import old_models
        model = getattr(old_models, 'MyModel')(inital_kernel_num = 32,dropout = 0.6,conv1kernel1 = 7,conv1kernel2 = 3)
    if model_name == 'Wacky':
        import old_models
        model = getattr(old_models, 'Wacky')(inital_kernel_num = 32,dropout = 0.6,conv1kernel1 = 7,conv1kernel2 = 3)
    print('parameters: ' +str(sum(p.numel() for p in model.parameters() if p.requires_grad)))
    if 'mayo' in model_name:
        model.train()
    else:
        model.eval()
    
    return model
def point(y_total,yhat_total,t):
    specificity = []
    sensitivity = []
    for i in tqdm(t):
        tn, fp, fn, tp = confusion_matrix(y_total, yhat_total>i).ravel()
        specificity.append( tn / (tn+fp) )
        sensitivity.append( tp / (tp+fn) )
    return t[(np.array(specificity) + np.array(sensitivity) - 1).argmax()]
def produce_df(model_name, test_loader, Project_name,folder = 'Test Results', depth = [1,2,2,3,3,3,3],channels = [32,16,24,40,80,112,192,320,1280],dilation = 1,stride = 2,expansion = 6,additional_inputs = 0,ds = None):
    try:
        os.mkdir(os.path.join(Project_name,folder))
    except:
        print("Folder exists")
    model = get_model(model_name,depth = depth,channels = channels,dilation = dilation,stride = stride,expansion = expansion, additional_inputs=additional_inputs)
    checkpoint = get_checkpoint(Project_name)
    checkpoint = torch.load(checkpoint)
    to_load = {'model': model}
    Checkpoint.load_objects(to_load=to_load, checkpoint=checkpoint)
    model.to('cpu')
    def thresholded_output_transform(yhat,y):
            y_pred, y = yhat,y
            y_pred = torch.sigmoid(y_pred)
            return y_pred, y
    yhat = torch.Tensor()
    if not ds is None:
        dl=torch.utils.data.DataLoader(ds, batch_size=1,num_workers=10,drop_last=False)
        for x,y in dl:
            count_ops(model,x)
            break
    if model_name == 'mayo' or model_name == 'mayo mortality':
        y_total = torch.LongTensor()
        yhat_total = torch.LongTensor()
    else:
        y_total = torch.Tensor()
        yhat_total = torch.Tensor()
    for x,y in tqdm(test_loader):
        yhat = model(x)
        if model_name == 'mayo':
            yhat,y = thresholded_output_transform(yhat,y)
        else:
            yhat,y = thresholded_output_transform(yhat,y)
        y_total = torch.cat((y_total,y),0)
        yhat_total = torch.cat((yhat_total,yhat.detach().cpu()),0)
    df = pd.DataFrame({'labels':y_total.flatten().tolist(),'Prediction':yhat_total.flatten().tolist()})
    df.to_csv(os.path.join(Project_name,folder,'Prediction.csv'))
    fpr,tpr, t = roc_curve(y_total,yhat_total)
    thresh = point(y_total,yhat_total,t)
    lw = 2
    cm(y_total.flatten(),yhat_total.flatten(),Project_name=Project_name,folder=folder,threshold = thresh)
    bootstrap(df,Project_name,folder)
    return yhat_total, y_total, fpr, tpr, t
def bootstrap(df,Project_name=None,folder=None):
    y_total,yhat_total = df['labels'],df['Prediction']
    fpr_boot = []
    tpr_boot = []
    aucs = []
    for i in tqdm(range(0,10000)):
        choices = np.random.choice(range(0,len(yhat_total)),int(len(yhat_total)/2))
        fpr,tpr, _ = roc_curve(y_total[choices],yhat_total[choices])
        fpr_boot.append(fpr)
        tpr_boot.append(tpr)
        aucs.append(auc(fpr,tpr))
    low,high = np.nanmean(aucs)-np.nanstd(aucs)*1.96,np.nanmean(aucs)+np.nanstd(aucs)*1.96
    lower_point = round(np.percentile(aucs,2.5),2)
    higher_point = round(np.percentile(aucs,97.5),2)
    mean_point = round(np.nanmean(aucs),2)
    x = plt.hist(aucs,bins = 50,label = 'mean: '+str(mean_point))

    plt.plot([np.percentile(aucs,2.5),np.percentile(aucs,2.5)],[0,max(x[0])],label = 'lower interval: '+str(lower_point))
    plt.plot([np.percentile(aucs,97.5),np.percentile(aucs,97.5)],[0,max(x[0])],label = 'higher interval: '+str(higher_point))
    plt.title("AUC Histogram")
    plt.xlabel("AUC")
    plt.legend()
    if not Project_name is None:
        plt.savefig(os.path.join(Project_name,folder,'AUC Histogram.png'))
    plt.show()
    
    plt.figure()
    lw = 2
    for i in range(0,1000):
        plt.plot(fpr_boot[i],tpr_boot[i], color='lightblue',
                 lw=lw)
    fpr,tpr, _ = roc_curve(y_total,yhat_total)
    plt.plot(fpr, tpr, color='darkorange',
             lw=lw, label='ROC curve (area = %0.2f)' % auc(fpr,tpr))
    plt.plot([0, 1], [0, 1], color='navy', lw=lw, linestyle='--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title('ROC Curve')
    plt.legend(loc="lower right")
    if not Project_name is None:
        plt.savefig(os.path.join(Project_name,folder,'ROC Curve.png'))
    plt.show()
def cm(y_total,yhat_total,Project_name = None,folder=None,threshold = 0.5):
    print(threshold)
    cm = confusion_matrix(y_total,yhat_total>threshold)
    tn, fp, fn, tp = confusion_matrix(y_total,yhat_total>threshold).ravel()
    specificity = ( tn / (tn+fp) )
    sensitivity= ( tp / (tp+fn) )
    print('Positive Predictive Value',round(tp/(tp+fp),2),'Negative Predictive Value', round(tn/(tn+fn),2), ' Specificty ', specificity, 'Sensitivity ', sensitivity)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=['No Event','Adverse Event'])
    disp.plot()
    plt.title("Confusion Matrix")
    if not Project_name is None:
        plt.savefig(os.path.join(Project_name,folder,'Confusion Matrix.png'))
    plt.show()
```
