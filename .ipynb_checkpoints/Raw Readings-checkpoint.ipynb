```python
def bootstrap(x,y):
    y_total,yhat_total = x,y
    aucs = []
    for i in range(0,40):
        choices = np.random.choice(range(0,len(yhat_total)),int(len(yhat_total)/2))
        fpr,tpr, _ = roc_curve(y_total[choices],yhat_total[choices])
        aucs.append(auc(fpr,tpr))
    lower_point = round(np.percentile(aucs,2.5),2)
    higher_point = round(np.percentile(aucs,97.5),2)
    fpr,tpr, _ = roc_curve(y_total,yhat_total)
    mean = auc(fpr,tpr)
    return mean, lower_point, higher_point
def produce_csv(folder,data_set,csv):
    model = models.EffNet(stride = 8, dilation = 2,channels = [32,16,24,40,80,112,192,320,1280,1867],hook=True)
    checkpoint = torch.load(folder)
    to_load = {'model': model}
    Checkpoint.load_objects(to_load=to_load, checkpoint=checkpoint)
    model.eval()
    root = data_set
    
    ds = EchoECG(root = root, csv = csv,model = 'RCRI_Net',rolling = 0,downsample=1,return_filename=True)
    bs = 300
    test_dataloader = torch.utils.data.DataLoader(ds, batch_size=bs,num_workers=40,shuffle = False,drop_last=False)
    y_total = torch.Tensor()
    yhat_total = torch.Tensor()
    fnames = []
    for x,y,fname in tqdm(test_dataloader):
        yhat = model(x)
        y_total = torch.cat((y_total,y),0)
        yhat_total = torch.cat((yhat_total,yhat.detach().cpu()),0)
        fnames.append(fname)
    return y_total,yhat_total,fnames
```

```python
folder = 'PhenoCode_with_PreOpNet/checkpointoptimizer: AdamW, lr: 0.004, /best_loss_model_19_val_loss=-0.0269.pt'
root = '/workspace/data/drives/Local_SSD/sdb/Processed_UKBB'
dic = {
    'normal':'/workspace/data/drives/Local_SSD/sdb/Processed_UKBB'
}
for i in dic.keys():

    csv = os.path.join('/workspace/data/drives/Local_SSD/sdd/data/',dic[i])
    y_total,yhat_total,fnames = produce_csv(folder,root,csv)
```
