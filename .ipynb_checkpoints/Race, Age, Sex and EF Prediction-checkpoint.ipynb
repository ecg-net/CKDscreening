```python
csv = pd.read_csv('/workspace/data/drives/Local_SSD/sdd/data/DL-Phenotype Test.csv')
quicky = csv[['Filename']]
quicky = quicky.merge(temp,on='Filename',how='left')
def race(x):
    if x =='CAUCASIAN':
        return 0
    if x =='BLACK':
        return 1
    else:
        return None
quicky['White_Or_Black'] = quicky.Race.apply(race)
quicky['Gender'] = 1*(quicky.Gender=='FEMALE')
quicky[['Filename','White_Or_Black']].dropna().to_csv('/workspace/data/drives/Local_SSD/sdd/data/White_Or_Black Test.csv',index=False)
quicky[['Filename','Gender']].dropna().to_csv('/workspace/data/drives/Local_SSD/sdd/data/Male_or_Female Test.csv',index=False)
```
