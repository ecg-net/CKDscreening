```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
total = pd.read_csv('total results 40 bootstrap.csv')
```

```python
total = total[~pd.isna(total.Mean)]
```

```python
labels = total['Group Labels'].value_counts().to_dict()
```

```python
data = []
text = []
for i in labels.keys():
    data.append(total[total['Group Labels']==i].Mean.dropna().to_list())
    text.append(i+'; count: '+str(labels[i]))
plt.figure(figsize=(15,10))
plt.boxplot(data,labels = text)
plt.xticks(rotation = -90)
plt.title("Total Results, all phenocodes")
plt.show()
```

```python
temp = total[total.positives>100]
labels = temp['Group Labels'].value_counts().to_dict()
data = []
text = []
for i in labels.keys():
    data.append(temp[temp['Group Labels']==i].Mean.dropna().to_list())
    text.append(i)
plt.figure(figsize=(15,10))
plt.boxplot(data,labels = text)
plt.xticks(rotation = -45)
plt.title("Total Results, phenocodes with 100 positive examples")
plt.show()
```

```python
temp = total[total.positives>100]
labels = temp['Group Labels'].value_counts()[:5].to_dict()
plt.figure(figsize=(15,10))
positions = np.array(range(0,5))
text = []

def set_box_color(bp, color):
    plt.setp(bp['boxes'], color=color)
    plt.setp(bp['whiskers'], color=color)
    plt.setp(bp['caps'], color=color)
    plt.setp(bp['medians'], color=color)

data = []
for i in labels.keys():
    data.append(temp[temp['Group Labels']==i]['Mean'].dropna().to_list())
    text.append(i+'; count: '+str(labels[i]))
    print(temp['Group Labels'].value_counts())
x = plt.boxplot(data,labels = text,positions = positions,widths = 0.1)
set_box_color(x, 'black') # colors are from http://colorbrewer2.org/

data = []
for i in labels.keys():
    data.append(temp[temp['Group Labels']==i]['qrs+'].dropna().to_list())
    print(temp['Group Labels'].value_counts())
y = plt.boxplot(data,positions = positions+0.15,widths = 0.1)
set_box_color(y, 'blue') # colors are from http://colorbrewer2.org/

data = []
for i in labels.keys():
    data.append(temp[temp['Group Labels']==i]['qrs-'].dropna().to_list())
    print(temp['Group Labels'].value_counts())
y = plt.boxplot(data,positions = positions+0.3,widths = 0.1)
set_box_color(y, 'red') # colors are from http://colorbrewer2.org/


# draw temporary red and blue lines and use them to create a legend
plt.plot([], c='black', label='Total')
plt.plot([], c='blue', label='QRS+')
plt.plot([], c='red', label='QRS-')
plt.legend()

plt.xticks(positions+0.15, text,rotation = -90)

plt.xticks()
plt.title("QRS Comparison phenocodes with 100 positive examples, top 5 most common groups")
plt.show()
```
